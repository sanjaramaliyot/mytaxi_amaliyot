```python
df['outcome'].replace(['recovering at home 03.03.2020','treated in an intensive care unit (14.02.2020)'],'recovering at home',inplace=True)
```

```python
df['outcome'].replace(['discharge','Discharged from hospital','Discharged'],'discharged',inplace=True)
```

```python
df['outcome'].replace(['critical condition, intubated as of 14.02.2020','critical condition'],'Critical condition',inplace=True)
```

```python
df['outcome'].replace(['Recovered'],'recovered',inplace=True)
```

```python
df['outcome'].replace(['severe illness'],'severe',inplace=True)
```

```python
df['outcome'].replace(['Recovered'],'recovered',inplace=True)
```

```python
df['outcome'].replace(['Symptoms only improved with cough. Currently hospitalized for follow-up.'],'Hospitalized',inplace=True)
```

```python
df['outcome'].replace(['Migrated_Other'],'Migrated',inplace=True)
```

```python
df['outcome'].replace(['stable','stable condition'],'Stable',inplace=True)
```
